```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('../data/heart.csv')
```

```python
df.info()
```

```python
# convert object columns to category
for col in df.select_dtypes(include = 'object').columns:
    df[col] = df[col].astype('category')
```

```python
# convert int64 columns to float64
for col in df.select_dtypes(include = 'int64').columns:
    df[col] = df[col].astype('float64')
```

```python
print(df.isnull().sum())
```

For numerical data, there are unrealistic entries. The minimum of RestingBP as 0 mm/Hg and cholesterol as 0 mm/dl.

```python
# Identify the entry with 0 resting blood pressure
df[df['RestingBP']==0]
```

```python
# Identify theentry with 0 as cholesterol
df[df['Cholesterol']==0]
```

- There is only one entry with 0 restingBP, possible treating method can be simply delete the entry, as it only consist of extremely small portion of the whole sample.

- While for 0 Cholesterol, there are 172 entries, deletion method is not applicable since it consist a large portion of the whole sample. 
We can use group-based median imputation to tackle this issue, as cholesterol level is one of the cause of hear disease clinically known. This method can preserve the relationship between cholesterol level with heart disease.
